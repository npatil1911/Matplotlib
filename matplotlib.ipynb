```python
import matplotlib.pyplot as plt
import numpy as np

x=['python','c','c++','java']
y=[85,70,60,82]
k=[20,50,30,19]

width=0.2
p=np.arange(len(x))
p1=[j+width for j in p]
plt.xlabel('Languages',fontsize=23)
plt.ylabel('Popularity in numbers',fontsize=23,)
plt.title('Graphical representation')

plt.xticks(p+width/2,x,rotation=40,fontsize=15)
plt.bar(p,y,width,color='r')
plt.bar(p1,k,width,color='y')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x=['python','c','c++','java']
y=[85,70,60,82]
k=[20,50,30,19]

width=0.2
p=np.arange(len(x))
p1=[j+width for j in p]
plt.xlabel('Languages',fontsize=23)
plt.ylabel('Popularity in numbers',fontsize=23,)
plt.title('Graphical representation')

plt.xticks(p+width/2,x,rotation=40,fontsize=15)
plt.barh(p,y,width,color='r')
plt.barh(p1,k,width,color='y')
plt.show()
```

# scatter plot

```python
x=[1,2,3,4,5,6]
y=[1,3,4,5,2,6]
color=('r','b','g','y','m','w')
size=[100,300,200,150,100,400]
plt.scatter(x,y,c=color,s=size)

plt.title('scatter plot')
plt.xlabel('day')
plt.ylabel('numbers')
```

```python
x=[1,2,3,4,5,6]
y=[1,3,4,5,8,6]
color=('r','b','g','y','m','w')
size=[600,800,850,150,100,400]
plt.scatter(x,y,c=color,s=size,marker='*',edgecolor='g',linewidth=2)

plt.title('scatter plot')
plt.xlabel('day')
plt.ylabel('numbers')
```

```python
x=[1,2,3,4,5,6]
y=[1,3,4,5,8,6]
m=[3,2,5,6,7,8]
color=('r','b','g','y','m','w')
size=[600,800,850,150,100,400]
plt.scatter(x,y,c=color,s=size,marker='*',edgecolor='g',linewidth=2)
plt.scatter(x,m,c=color,s=size,marker='*',edgecolor='g',linewidth=2)
plt.colorbar()
plt.title('scatter plot')
plt.xlabel('day')
plt.ylabel('numbers')
```

# Fill_Between

```python
x=[1,2,3,4,5]
y=[1,2,3,4,5]

plt.plot(x,y,color="k")

plt.fill_between(x,y,color="g")
plt.title('python')
plt.xlabel('x-axis')
plt.ylabel('y-axis')

plt.show()
```

```python
import numpy as np
x=np.array([1,2,3,4,5])
y=np.array([1,2,3,4,5])

plt.plot(x,y,color="g")

plt.fill_between(x,y,color="m",where=(x>2) & (x<=4),alpha=0.5)
plt.title('python')
plt.xlabel('x-axis')
plt.ylabel('y-axis')

plt.show()
```
